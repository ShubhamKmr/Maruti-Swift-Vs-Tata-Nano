# Image Classification of the maruti swift car and tata nano car using keras library.

```python
#colors tuple consists of these colors of the car in the image.
colors = ['white ','red ','grey ','black ','yellow ','blue ','orange ']
```

```python
#model tuple consist of the model of car to be classified
model = ['nano car','swift car']
```

```python
#You need to download google_images_download library which is used too download max 100 image max for a particular keyword searched
#you can download it by going to cmd and typing the command pip install google_images_download
from google_images_download import google_images_download   #importing the library
response = google_images_download.googleimagesdownload()   #class instantiation
#Applying loop to download for combination of color and cars
for car in model:
    for color in colors:
        arguments = {"keywords":color + car,"limit":100,"print_urls":True}   #creating list of arguments
        paths = response.download(arguments)   #passing the arguments to the function
        print(paths)
```

```python
#Importing all important libraies 
from keras.models import Sequential 
from keras.layers import Convolution2D
from keras.layers import MaxPooling2D
from keras.layers import Flatten
from keras.layers import Dense
from keras.layers import Dropout
import warnings
warnings.filterwarnings('ignore')
```

```python
#Sequential is used to initialize the Deep learning model
classifier = Sequential()
```

```python
#1st CNN layer 
'''
Applying Convolution to catch the different features in the image.Using 32 different feature detector of shape 3X3 that means 32 different 
feature maps are produced.Also taking the common size for the image 128X128 pixel 3 -> means the images are colored,Activation function
is retifier linear unit used to increase non-linearity in the image
'''
classifier.add(Convolution2D(32,3,3,input_shape=(128,128,3),activation='relu'))
```

```python
#Pooling is carried out to find the correlation between the pixel even if the image are not in it normal orientation.using pool filter of 2X2 
classifier.add(MaxPooling2D(pool_size=(2,2)))
```

```python
#2nd CNN layer
classifier.add(Convolution2D(32,3,3,activation='relu'))
classifier.add(MaxPooling2D(pool_size=(2,2)))
```

```python
#3rd CNN layer
classifier.add(Convolution2D(32,3,3,activation='relu'))
classifier.add(MaxPooling2D(pool_size=(2,2)))
```

```python
#Flatten is used to produced single dimension input to the ANN.
classifier.add(Flatten())
```

```python
#It is the 1st hidden layer having 128 neurons or node with activation function as reLU.
classifier.add(Dense(output_dim=128,activation='relu'))
```

```python
#Dropout are basically applied to reduce overfitting
classifier.add(Dropout(0.2))
```

```python
#It is the 2nd hidden layer having 128 neurons or node with activation function as reLU.
classifier.add(Dense(output_dim=128,activation='relu'))
```

```python
#Dropout are basically applied to reduce overfitting
classifier.add(Dropout(0.2))
```

```python
#This the output layer having probablistic output by using sigmoid function as activation function
classifier.add(Dense(output_dim=1,activation='sigmoid'))
```

```python
#Finally Compilation to the network build using Sochastic Gradient Descent,having 2 possible output.So using binary_crossentropy as loss function
classifier.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
#Now going for Image Augmentation technique
from keras.preprocessing.image import ImageDataGenerator 
```

```python
#I did not had much images So, applying shear_range, zoom_range, horizontal_flip to get more no to different image from the existing once.
train_datagen = ImageDataGenerator(rescale=1/255,shear_range=0.2,zoom_range=0.2,horizontal_flip=True)
```

```python
#Just rescaling the testing images
test_datagen = ImageDataGenerator(rescale=1/255)
```

```python
#Now providing the path of the Image Dataset making batch of 32 ie after 32 images the weights will be updated target_size = input image size
training_set = train_datagen.flow_from_directory(directory='Dataset/Train',batch_size=32,target_size=(128,128),class_mode='binary')
```

```python
#Now providing the path of the Image Dataset making batch of 32 ie after 32 images the weights will be updated target_size = input image size
testing_set = test_datagen.flow_from_directory(directory='Dataset/Test',batch_size=32,class_mode='binary',target_size=(128,128))
```

```python
#It is used to train as well as predict the model together.
classifier.fit_generator(generator=training_set,samples_per_epoch=737,nb_epoch=25,validation_data=testing_set,nb_val_samples=184)
```
